```python
# must install chembl_websource_client using pip

import pandas as pd
from chembl_webresource_client.new_client import new_client as nc
```

```python
target_query = nc.target.search("acetylcholinesterase")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[0]
selected_target
```

```python
res = nc.activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.standard_type.unique()
```

```python
df.shape
```

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df2.canonical_smiles.notna()]
```

```python
df2.shape
```

```python
df2
```

```python
df3 = df2.drop_duplicates(['canonical_smiles'])
df3
```

```python
# Handling Missing Data


bioactivity_class = []
for i in df3.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")

selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df4 = df3[selection]
df4
```

```python
df4.to_csv('bioactivity_preprocessed_data.csv', index=False)
df5 = pd.read_csv("bioactivity_preprocessed_data.csv")
df5
```

```python
df6 = pd.concat([df5,pd.Series(bioactivity_class, name="bioactivity_class")], axis=1)
df6
```

```python
df6.to_csv('bioactivity_preprocessed_data.csv', index=False)
```
