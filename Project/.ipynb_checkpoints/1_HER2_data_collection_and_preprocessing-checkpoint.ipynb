```python
# must install chembl_websource_client using pip

import pandas as pd
from chembl_webresource_client.new_client import new_client as nc
```

```python
target_query = nc.target.search("HER2")
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

```python
res = nc.activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
df = pd.DataFrame.from_dict(res)
```

```python
df
```

```python
df.standard_type.unique()
```

```python
df.shape
```

```python
df2 = df[df.standard_value.notna()]
df3 = df2[df2.canonical_smiles.notna()]
```

```python
df3.shape
```

```python
df3
```

```python
df4 = df3.drop_duplicates(['canonical_smiles'])
df4
```

```python
selection = ['molecule_chembl_id', 'canonical_smiles', 'standard_value']
df5 = df4[selection]
```

```python
df5.to_csv('bioactivity_preprocessed_data.csv', index=False)
```

```python
df6 = pd.read_csv("bioactivity_preprocessed_data.csv")

df6
```

```python
df6[df6.canonical_smiles == 0]
```

```python
# Handling Missing Data

bioactivity_class = []
for i in df6.standard_value:
  if float(i) >= 3000:
    bioactivity_class.append("inactive")
  elif float(i) <= 200:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
df7 = pd.concat([df6,pd.Series(bioactivity_class, name="bioactivity_class")], axis=1)
df7
```

```python
counts = df7['bioactivity_class'].value_counts()
counts
```

```python
df7.to_csv('bioactivity_data.csv', index=False)
```
