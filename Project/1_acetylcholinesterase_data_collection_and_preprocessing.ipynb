```python
# Handling Missing Data

bioactivity_class = []
for i in df8.standard_value:
  if float(i) >= 10000:
    bioactivity_class.append("inactive")
  elif float(i) <= 1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
df9 = pd.concat([df8,pd.Series(bioactivity_class, name="bioactivity_class")], axis=1)
df9
```

```python
df9.to_csv('bioactivity_data.csv', index=False)
```
